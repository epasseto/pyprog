Aqui eu crio minha primeira função. Ela testa se nosso dataframe possui algum **registro duplicado**:

*de fato eu poderia continuar escrendo apenas scripts... mas se isso é uma linguagem de programação completa, por que não explorar recursos mais avançados?*

```python
def tem_duplicados(df):
    if df.duplicated().any():
        print("Possui {0} valores duplicados".format(df.duplicated().sum()))
    else:
        print("Não possui valores duplicados")
```

Chamando minha função:

```python
tem_duplicados(dfInfracao)
```

Quem faz sua primeira função, faz qualquer uma... criando a partir daquela minha primeira função, uma outra função para saber se todos os valores de PENALID **são únicos**:

```python
def sao_unicos(df):
    if df["PENALID"].is_unique:
        print("OK, São valores únicos")
    else:
        print("ATENÇÃO: Não são valores únicos!")
```

Chamando minha nova função:

```python
sao_unicos(dfInfracao)
```

Epa, eu quero **generalizar** essa minha nova função:

```python
def sao_unicosg(df, campo):
    if df[campo].is_unique:
        print("OK, São valores únicos")
    else:
        print("ATENÇÃO: Não são valores únicos!")
```

Eu sei que este campo aqui **se repete**, a resposta consistente seria que não são valores únicos:

```python
filtrar = 'PENALID'

sao_unicosg(dfInfracao, filtrar)
```

Se houver um ID repetido, isso sim seria um **problema**. Então minha melhor resposta seria que são valores únicos:

```python
filtrar = 'ID_OCORR'

sao_unicosg(dfInfracao, filtrar)
```

```python
NovosNomes = ['ID_OCO', 'número', 'ano', 'penalidade', 'documento', 'processo', 
              'usuário', 'ID_USU', 'finalidade', 'resolução', 'rio']
```
