**Deslocamento** versus **Velocidade**

Se eu me desloco no tempo sempre para **frente**, eu não posso ter uma velocidade **negativa**, portanto a curva vermelha está excluída. Reparando novamente, são dois momentos em que a velocidade parece estar zero, aumenta e depois volta a ficar zero. E os dois são relativamente da mesma grandeza. Então a linha verde será a correta

**Deslocamento** versus **Aceleração**

O mais adequado é fazer isso em duas etapas. Então, para **velocidade**, eu estou começando parado, daí acelero relativamente devagar, até chegar ao ponto zero e desacelero no mesmo padrão, até passar o ponto zero até um ponto X simétrico. Bom, nesse trecho para **aceleração**, eu estou acelerando para **frente**, pois não possuía velocidade nenhuma e preciso cumprir terreno apenas para a frente. Então minha aceleração é **positiva** até atingir o ponto zero e depois **negativa**, pois devo agora desacelerar da mesma maneira como acelerei até chegar no meu ponto X simétrico. Então tanto a linha rosa como a vermelha servem

No segundo trecho, também em duas etapas. Para **velocidade** eu devo engatar uma marcha à ré a fim de voltar ao meu ponto zero, então eu devo aplicar uma **aceleração** negativa. E rápida, pois eu consumo muito pouco tempo para muito deslocamento! E depois eu preciso dar uma aceleração **positiva** para poder zerar minha velocidade **negativa**, até eu parar, sobre o meu ponto zero. Bom, o que descreve essas curvas agora é apenas a linha vermelha

Se eu tivesse derivada **terceira** e assim em diante, o melhor seria primeiro resolver a derivada primeira, depois a a segunda e provavelmente verificar apenas o **sinal** da minha última derivada, que nesse caso seria a terceira. É normalmente desta maneira que os físicos procedem. E é claro, verificar o **domínio** da minha função para ignorar aquelas soluções fora do domínio (por exemplo, uma viga de aço com massa **negativa**)
