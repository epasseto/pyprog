#### Anatomia do Matplotlib

---

O **Matplotlib**, que dá toda a base do **Pyplot** é todo modular e os módulos podem possuir **submódulos**, depois **funções** e **métodos**. Uma **função** do Python é uma **classe** e tem uma forma geral da seguinte maneira **.figure()**. Quando ela é chamada, ela cria um **objeto** em sua memória.

O **objeto**, de uma maneira geral, não faz nada. Mas ele pode ser **chamado** e receber **parâmetros** e ele irá fazer as tarefas que você mandar. Um **objeto** é como se fosse um ninho, mas o que nos interessam são os ovos...

---

Em geral, no Python existem duas maneiras de se fazer as coisas. A maioria dos usuários quer programar de uma maneira elegante e rápida e não quer ficar se preocupando com classes e objetos. E então eles usam a codificação **implícita**. Você chama direto a classe e ela cria para ti o objeto. E depois você pode chamar a classe e ir elaborando seu objeto, até chegar no caso, ao comando de plotar tudo isso. Se você não quer muito trabalho, ou se quer usar coisas mais básicas, é um caminho perfeito!

No entanto, em alguns casos isso não atende. Então eu passo à codificação **explícita**.

Na codificação **explícita** primeiro eu crio um objeto com o nome **figura**, que não possui nada. Ele serve apenas de infraestrutura para o resto.

---

Se eu quiser que as duas plotagens ocorram na mesma caixa, eu indico algo como (121) e (121). Se eu usar algo como (151) e (155), ele preparará cinco coluas e colocará o primeiro gráfico na colna 1 e o segundo gráfico na coluna 5, os demais ficando em branco: 

    add_subplot(111)

#### Fill between

---

Uma das funções legais de explorar é a **.fill_between(X, Y1, Y2=0)**. Ela pode gerar ótimos resultados em gráficos. Uma barra pode ser criada com este comando.

São três parâmetros:

 - X - é uma matriz de tamanho N, contendo as coordenadas X dos nós definindo as curvas

 - Y1 - é uma matriz de tamanho N ou escalar, definindo as coordenadas X dos nós definindo a primeira curva
 
 - Y2 - a mesma coisa, padrão=0, dos nós definindo a segunda curva

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 2, 0.01)
y1 = np.sin(2*np.pi*x)
y2 = 1.2*np.sin(4*np.pi*x)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)

ax1.fill_between(x, 0, y1)
ax1.set_ylabel('entre y1 e 0')

ax2.fill_between(x, 0, y1)
ax2.set_ylabel('entre y1 e 1')

ax3.fill_between(x, y1, y2)
ax3.set_ylabel('entre y1 e y2')
ax3.set_xlabel('x')

# now fill between y1 and y2 where a logical condition is met.  Note
# this is different than calling
#   fill_between(x[where], y1[where],y2[where]
# because of edge effects over multiple contiguous regions.
fig, (ax, ax1) = plt.subplots(2, 1, sharex=True)
ax.plot(x, y1, x, y2, color='black')
ax.fill_between(x, y1, y2, where=y2 >= y1, facecolor='green', interpolate=True)
ax.fill_between(x, y1, y2, where=y2 <= y1, facecolor='red', interpolate=True)
ax.set_title('fill between com where')

# Test support for masked arrays.
y2 = np.ma.masked_greater(y2, 1.0)
ax1.plot(x, y1, x, y2, color='black')
ax1.fill_between(x, y1, y2, where=y2 >= y1, facecolor='green', interpolate=True)
ax1.fill_between(x, y1, y2, where=y2 <= y1, facecolor='red', interpolate=True)
ax1.set_title('Agora regiões com y2>1 recebem a máscara')

# This example illustrates a problem; because of the data
# gridding, there are undesired unfilled triangles at the crossover
# points.  A brute-force solution would be to interpolate all
# arrays to a very fine grid before plotting.

# show how to use transforms to create axes spans where a certain condition is satisfied
fig, ax = plt.subplots()
y = np.sin(4*np.pi*x)
ax.plot(x, y, color='black')

# use the data coordinates for the x-axis and the axes coordinates for the y-axis
import matplotlib.transforms as mtransforms
trans = mtransforms.blended_transform_factory(ax.transData, ax.transAxes)
theta = 0.9
ax.axhline(theta, color='green', lw=2, alpha=0.5)
ax.axhline(-theta, color='red', lw=2, alpha=0.5)
ax.fill_between(x, 0, 1, where=y > theta, facecolor='green', alpha=0.5, transform=trans)
ax.fill_between(x, 0, 1, where=y < -theta, facecolor='red', alpha=0.5, transform=trans)

plt.show()
```

**Apagar um eixo**

---

Quando se trabalha com codificação **explícita** no Matplotlib, é possível fazer muitas coisas. Uma delas é apagar um eixo já definido:

```python
# Import `pyplot` from `matplotlib`
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("ggplot")

# Initialize the plot
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

# Plot the data
ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])
ax2.axhline(0.45)
ax1.axvline(0.65)
ax3.scatter(np.linspace(0, 1, 5), np.linspace(0, 5, 5))
ax2.legend(["alô mamãe", "oie"], bbox_to_anchor=(1.5, 0.9))

# Delete `ax3`
fig.delaxes(ax3)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)


def plot_scatter(ax, prng, nb_samples=100):
    """Scatter plot.
    """
    for mu, sigma, marker in [(-.5, 0.75, 'o'), (0.75, 1., 's')]:
        x, y = prng.normal(loc=mu, scale=sigma, size=(2, nb_samples))
        ax.plot(x, y, ls='none', marker=marker)
    ax.set_xlabel('X-label')
    return ax


def plot_colored_sinusoidal_lines(ax):
    """Plot sinusoidal lines with colors following the style color cycle.
    """
    L = 2 * np.pi
    x = np.linspace(0, L)
    nb_colors = len(plt.rcParams['axes.prop_cycle'])
    shift = np.linspace(0, L, nb_colors, endpoint=False)
    for s in shift:
        ax.plot(x, np.sin(x + s), '-')
    ax.set_xlim([x[0], x[-1]])
    return ax


def plot_bar_graphs(ax, prng, min_value=5, max_value=25, nb_samples=5):
    """Plot two bar graphs side by side, with letters as x-tick labels.
    """
    x = np.arange(nb_samples)
    ya, yb = prng.randint(min_value, max_value, size=(2, nb_samples))
    width = 0.25
    ax.bar(x, ya, width)
    ax.bar(x + width, yb, width, color='C2')
    ax.set_xticks(x + width)
    ax.set_xticklabels(['a', 'b', 'c', 'd', 'e'])
    return ax


def plot_colored_circles(ax, prng, nb_samples=15):
    """Plot circle patches.

    NB: draws a fixed amount of samples, rather than using the length of
    the color cycle, because different styles may have different numbers
    of colors.
    """
    for sty_dict, j in zip(plt.rcParams['axes.prop_cycle'], range(nb_samples)):
        ax.add_patch(plt.Circle(prng.normal(scale=3, size=2),
                                radius=1.0, color=sty_dict['color']))
    # Force the limits to be the same across the styles (because different
    # styles may have different numbers of available colors).
    ax.set_xlim([-4, 8])
    ax.set_ylim([-5, 6])
    ax.set_aspect('equal', adjustable='box')  # to plot circles as circles
    return ax


def plot_image_and_patch(ax, prng, size=(20, 20)):
    """Plot an image with random values and superimpose a circular patch.
    """
    values = prng.random_sample(size=size)
    ax.imshow(values, interpolation='none')
    c = plt.Circle((5, 5), radius=5, label='patch')
    ax.add_patch(c)
    # Remove ticks
    ax.set_xticks([])
    ax.set_yticks([])


def plot_histograms(ax, prng, nb_samples=10000):
    """Plot 4 histograms and a text annotation.
    """
    params = ((10, 10), (4, 12), (50, 12), (6, 55))
    for a, b in params:
        values = prng.beta(a, b, size=nb_samples)
        ax.hist(values, histtype="stepfilled", bins=30,
                alpha=0.8, density=True)
    # Add a small annotation.
    ax.annotate('Annotation', xy=(0.25, 4.25), xycoords='data',
                xytext=(0.9, 0.9), textcoords='axes fraction',
                va="top", ha="right",
                bbox=dict(boxstyle="round", alpha=0.2),
                arrowprops=dict(
                          arrowstyle="->",
                          connectionstyle="angle,angleA=-95,angleB=35,rad=10"),
                )
    return ax


def plot_figure(style_label=""):
    """Setup and plot the demonstration figure with a given style.
    """
    # Use a dedicated RandomState instance to draw the same "random" values
    # across the different figures.
    prng = np.random.RandomState(96917002)

    # Tweak the figure size to be better suited for a row of numerous plots:
    # double the width and halve the height. NB: use relative changes because
    # some styles may have a figure size different from the default one.
    (fig_width, fig_height) = plt.rcParams['figure.figsize']
    fig_size = [fig_width * 2, fig_height / 2]

    fig, axes = plt.subplots(ncols=6, nrows=1, num=style_label,
                             figsize=fig_size, squeeze=True)
    axes[0].set_ylabel(style_label)

    plot_scatter(axes[0], prng)
    plot_image_and_patch(axes[1], prng)
    plot_bar_graphs(axes[2], prng)
    plot_colored_circles(axes[3], prng)
    plot_colored_sinusoidal_lines(axes[4])
    plot_histograms(axes[5], prng)

    fig.tight_layout()

    return fig

if __name__ == "__main__":

    # Setup a list of all available styles, in alphabetical order but
    # the `default` and `classic` ones, which will be forced resp. in
    # first and second position.
    style_list = ['default', 'classic'] + sorted(
        style for style in plt.style.available if style != 'classic')

    # Plot a demonstration figure for every available style sheet.
    for style_label in style_list:
        with plt.style.context(style_label):
            fig = plot_figure(style_label=style_label)

    plt.show()
```

## 1. uma **relação**?
Gráficos do **tipo I - relações**

**Relações** servem para podermos tentar compreender possíveis variações comuns entre duas ou mais variáveis. Pontos muito espalhados e caóticos podem demonstrar relações **fracas** ou a inexistência delas e pontos mais agrupados podem sugerir relações **fortes**. As relações podem nos contar outras coisas, como uma possível relação **direta** ou **inversa** entre duas grandezas, ou uma tendência a que valores muito elevados em notas de estudantes tendam a não se repetir, mesmo naquela mesma matéria. Elas podem nos contar como a doçura de um vinho pode alterar a nota para **mais** se a doçura vier acompanhada por teor alcoólico mais elevado, mas que pode alterar a nota para **menos** em se tratando de vinhos da região Norte de Portugal. Às vezes essas relações são realmente complexas. Mesmo assim vale a pena observar com atenção os gráficos de relações para entender a sutileza destas relações. 

entre **duas grandezas**

 -  **Scatter Plot** - normalmente se procura por uma tendência, como uma **reta**, **curva** ou outra **função**, de como uma se altera em função da outra. Quando se obtém essa curva ou função, faz-se seu **ajuste** e ela representará matematicamente essa tendência. Isso é útil em racionalização de processos. Às vezes as variações são mais sutis e cabe à criatividade do cientista de dados tentar explorar essas nuances. **Outliers** também podem ser estudados à parte. É útil por exemplo para verificar o quanto o teor alcoólico de amostras de vinho pode estar relacionado a notas mais altas ou mais baixas em uma degustação

entre **três ou mais grandezas**

 - **Scatter Plot a ou Bubble Saze** - o **peso** de uma grandeza ganha uma dimensão extra no diâmetro das bolas. Isso em função de duas variáveis, que são consideradas **fundamentais**. Primeiro se sugere a escolha dessas duas grandezas e depois se vai ajustando o resto do gráfico. Se elas não nos atendem muito bem, começamos de novo. Não queremos nuvens homogêneas de pontos, elas normalmente nos dizem muito pouco. É uma exploração muito útil para induzir **novas hipóteses** sobre possíveis relações complexas, como o tamanho da população de um país, colorido por regiões, versus PIB e expectativa de vida

## 3. uma **distribuição**?
Gráficos do **tipo III - distribuição**

**Distribuições** são como a queda de maçãs no solo. A macieira carregada, a maior densidade de maçãs perto do tronco, as frutas começam a amadurecer e caem ao solo. ### Anatomia do Seaborn

---

#### Nível de Eixo (axis level)

    regplot()
    
    boxplot()
    
    kdeplot()

    ...

Todas elas pegam um argumento explícito **ax** e retornam um objeto **Ax**

E se eu defino por exemplo:

    sns.boxplot(x="gorjeta_total", data=tips)
    
O sistema me retorna um objeto **Axes Subplot** definido pelo Matplotlib:

    <matplotlib.axes._subplots.AxesSubplot object ...
    
---

#### Nível de Figura (figure level)

    lmplot()
    
    factorplot()
    
    jointplot()
    
    ...

E se eu defino por exemplo:

    sns.lmplot('x', 'y', data, size=7, truncate=True, scatter_kws={"s": 100})

Agora o sistema me retorna um objeto **Facet Grid** definido pelo Seaborn:

    <seaborn.axisgrid.FacetGrid object ...
    
O FacetGrid por sua vez está dividido entre argumentos de:

- FacetGrid.fig

- FacetGrid.axes

*Estratégia geral: prefira customizar em nível de **Eixo** (por exemplo no **.lmplot**) para manter a homogeneidade geral dos seus gráficos. Depois faça os ajustes finos em nível de **Figura** (por exemplo passando parâmetros para o **.regplot()**). Parece complicado, mas é só imaginar que se está trabalhando com uma escada de dois degraus!*

----

E estas **Figurs** criam gráficos que estarão montadas sobre **Eixo**, organizados de uma maneira intuitiva

As funções de nível de **Figura** terão controle total sobre quela figura e portanto, você não conseguirá plotar uma **.lmplot()** sobre uma já existente!

Quando você chama uma funções de nível de figura, você sempre está:

- inicializando a figura

- a setando para a plotagem específica que ela irá fazer

*Lembre-se de que tudo isso é baseado em **Orientação a Objeto** e assim, quando você inicializa sua figura, você está materializando uma **Classe** em um **Objeto**. Uma vez feito isso, ele ocupará espaço na memória e se comportará como um ser específico!

Em nível de **Figura**, você pode acessar o objeto **Axes** através do argumento **axes**, setando as propriedades **xlim** e **ylim**

```python
# Import the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
import pandas as pd

# Initialize the data
x = 10 ** np.arange(1, 10)
y = x * 2
data = pd.DataFrame(data={'x': x, 'y': y})

# Create lmplot
lm = sns.lmplot('x', 'y', data, height=4, truncate=True, scatter_kws={"s": 100})

# Get hold of the `Axes` objects
axes = lm.ax

# Tweak the `Axes` properties
axes.set_ylim(-1000000000,) # <--AQUI!

# Show the plot
plt.show()
```

A partir de uma rede **Barabási-Albert**, a função **ego_graph()** produz uma **Egonet**

O modelo ** gera redes **sem escala** e aleatórias. Novos nós podem ser inseridos ao longo do tempo e esses podem se ligar a outros pré-existentes. Isso é governado por uma distribuição de grau tipo **lei da potência**

Imagine aqui que este é meu perfil **Facebook**. Eu estou ligado a outras pessoas que por sua vez se ligam entre si:

```python
from operator import itemgetter
import networkx as nx
import matplotlib.pyplot as plt

if __name__ == '__main__':
    # Create a BA model graph
    n=1000
    m=2
    G=nx.generators.barabasi_albert_graph(n,m)
    
    # find node with largest degree
    node_and_degree=G.degree()
    (largest_hub,degree)=sorted(node_and_degree,key=itemgetter(1))[-1]

    # Create ego graph of main hub
    hub_ego=nx.ego_graph(G,largest_hub)

    # Draw graph
    pos=nx.spring_layout(hub_ego)
    nx.draw(hub_ego,pos,node_color='b',node_size=50,with_labels=False)

    # Draw ego as large and red
    nx.draw_networkx_nodes(hub_ego,pos,nodelist=[largest_hub],node_size=300,node_color='r')
    plt.savefig('ego_graph.png')
    plt.show()
```

Este não informa qual registrador precisa ser carregado e qual porção da memória alocada para esta operação. Então o seu código depois de terminado passa por um processo de **linkagem**  (banco de dados) e **compilação** para ser transformado em linguagem de máquina

## Explicação sobre **Método**, **Função** e **Parâmetro** no Python

---

**Método Python** encontrado [aqui](https://www.geeksforgeeks.org/difference-method-function-python)

- é chamado por seu **nome**, mas está associado a um **objeto** (é dependende de um objeto)

- um método é passado **implicitamente** para o objeto que é invocado

- ele pode **ou não** retornar algum dado

- o método pode operar sobre o dado (variáveis de **instância**) que estiver contida na sua classe correspondente

**Função Python**

- também é chamada por seu **nome** (mas é um bloco de código independente)

- pode ter diversos **parâmetros**, ou nenhum. Se nenhum **parâmetro** for passado, eles são passados **explicitamente**

- pode **ou não** retornar algum dado

- o mais importante é que funções não lidam com o conceito de **Classe** e **Instância**!

Essa explicação resolve o problema, mas é um pouco enrolada

Primeiro temos que saber um pouco sobre **Orientação a Objeto**

O Python, como todas as linguagens modernas de programação é orientado a objeto. A ideia de objetos surgiu no final década de 1970. Os computadores começaram a ficar cada vez mais avançados e baratos e podiam ser ligados até em televisores preto e branco. Se começou a perceber que o potencial da computação não estaria apenas em grandes centros, mas talvez até na casa de pessoas. 
## Decorators em Python [aqui](https://pythonhelp.wordpress.com/2013/06/09/entendendo-os-decorators/)

---

É uma maneira de adicionar funcionalidades às funções/métodos/classes, sem mudar seus códigos. Em programação orientada a objeto, chama-se **Sobrever**

Às vezes um **Framework** já traz uma coleção de **Decoradores** prontos. Nestes casos, basta usar o **Decorador** e o código já foi escrito para você! O pessoal diz que eles são **açucar sintático** para o programador

#### Implementar um Decorador

É uma função que recebe uma função como parâmetro e faz **algo** e então executa a função-parâmetro e retorna o resultado

O **algo** é a funcionalidade que adicionamos à nossa função original através do **Decorador**

Na verdade, eu posso colocar **Decorators** em qualquer coisa do Python, como botões, caixas, consultas e outras funções. É como customizar um fusca: vale tudo!

---

Suponha que eu tenha uma função qualquer:

```python
def dobro(x):
    return x*2
 
dobro(10)
```

OK, minha função opera, tudo certo. Mas em alguns casos, eu quero que **todas** as funções do meu programa escrevam o nome da função a ser executada. Por exemplo, em um modo de **segurança**

---

A primeira coisa que eu coloco aqui é uma linha antes da minha função chamando o **Decorador**. É como chamar a **moldura** ao pedir uma **pintura**:

```python
def nomeia_func(func):# <--recebe uma função func
    def finterna(*args, **kwargs):
        print ("Chamando função: {}".format(func.__name__))
        return func(*args, **kwargs)# <-- executa e retorna o valor de retorno da func
    return finterna# <--retorna uma função finterna

@nomeia_func# <-- meu decorador
def dobro(x):
    return x*2
 
dobro(10)
```

A **finterna** é um objeto como qualquer outro. Só que este objeto é uma **Função**. Mas sendo um objeto, a função é **instanciada**, recebe um nome (finterna) e pode ser retornada, como qualquer objeto. Tudo isso sem ser **executada**, pois não chamamos a **finterna**!

A palavra chave **def** apenas **cria** a função, sem a executar. Então **nomeia_func** cria uma função, configura ela para executar **func()** ao seu final e não executa **nada**! O objeto função é retornado e então está pronto para ser chamado por quem o receber

---

#### Conceito de closures [aqui](https://developer.mozilla.org/pt-BR/docs/Web/JavaScript/Guide/Closures)

**Closure** é uma função **aninhada**, definda dentro de outra função maior. Então ela fica disponível apenas dentro do corpo daquela função. Ela não possui variáveis próprias locais e então reusa as variáveis declaradas na função pai. Ela permite você associar dados do ambiente com uma função que trabalha esses dados

---

Pareceu maluco, não? Mas no fundo é tudo bastante **simples**. O Python é uma linguagem interpretada e linguagens interpretadas possuem a fama de serem **toscas**. Toscas porque são lentas e não possibilitam muita coisa. Lembra o antigo BASIC dos TK-83? Então!

Como superar essa maldição? A única maneira é em algum momento haver um botãozinho escrito "compile o código a partir daqui". Esse botãozinho no Python é **óbvio** que existe, mas ele ficou disfarçado. Então eu vou passar alguma instrução para o Python, eu chamo um **Método**. O **Método** em si não faz nada. Ele só é criado (como um **Summon** de um lobo nível 1, por um mago, em jogo de RPG) e passa a ocupar um lugar na memória da minha máquina

Pronto! Esse cara aí é um **Objeto**! E a estrutura interna do meu **Objeto** é sim, **compilada!**. Daí eu crio o objeto **máquina de lavar roupa**. Ela sai lavando roupa feito doida? Claro que não. Ela fica lá paradinha no canto dela, preparada para lavar o que eu meter lá dentro. É a **mesma** coisa aqui!

O que o aquela minha **nomeia_função** faz é pegar a função interna que foi passada para ela pela variável **func** e daí, na função **aninhada**, botar um **print**, meter a execução da própria função recebida, empacotar tudo e devolver uma **finterna** na maneira de um objeto, compiladinho, bonitinho para ser executado.

Daí quando eu chamo minha função **.dobro()**, aí sim, ele executa esse troço aí de cima!

---

E esse negócio de...

    *args - usado para passar um número variável de argumentos para uma função
    
    dobro('hello', 10, 'raio')
    
    **kwargs - mesma coisa, mas da maneira de dicionário
    
    dobro(entrada = 10, argumento = 'mediana')
    
Sabe por que isso aparece no meu **Decorador**? Simplesmente porque um dos princípios básicos da programação é escrever código da maneira mais **genérica** e **universal** possível. Código escrito assim é facilmente **reaproveitável**!

Este é um princípio da **Epistemologia**, adaptado à **Ciência da Computação**. Então a operação **raiz quadrada** funciona tão bem na Terra como na Lua, ou embaixo d´água! Quem duvida disso, deveria procurar na biblioteca o livro de Karl Popper, *Conhecimento Objetivo*

Então quando eu escrevo esse meu Decorador, eu lá sei quais os **parâmetros** e o número deles que será passado para minha função a ser decorada? Não! Meu caso foi ridículo, pois a única função testada pega apenas um argumento:

    dobro(10)
    
Mas e se fosse como nos exemplos acima? O a **mistura** deles:

    dobro(entrada = 10, 3.1416, 'raio')

Então esses dois passam um rodo em tudo isso e minha função enfeitada fica OK! E se eu não fizer isso? Poderei ter **sérios problemas futuros**!

**Plotting a diagonal correlation matrix**

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")

# Generate a large random dataset
rs = np.random.RandomState(33)
d = pd.DataFrame(data=rs.normal(size=(100, 26)),
                 columns=list(ascii_letters[26:]))

# Compute the correlation matrix
corr = d.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(9, 7))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```

**Discovering structure in heatmap data**

```python
import pandas as pd
import seaborn as sns
sns.set()

# Load the brain networks example dataset
df = sns.load_dataset("brain_networks", header=[0, 1, 2], index_col=0)

# Select a subset of the networks
used_networks = [1, 5, 6, 7, 8, 12, 13, 17]
used_columns = (df.columns.get_level_values("network")
                          .astype(int)
                          .isin(used_networks))
df = df.loc[:, used_columns]

# Create a categorical palette to identify the networks
network_pal = sns.husl_palette(8, s=.45)
network_lut = dict(zip(map(str, used_networks), network_pal))

# Convert the palette to vectors that will be drawn on the side of the matrix
networks = df.columns.get_level_values("network")
network_colors = pd.Series(networks, index=df.columns).map(network_lut)

# Draw the full plot
sns.clustermap(df.corr(), center=0, cmap="vlag",
               row_colors=network_colors, col_colors=network_colors,
               linewidths=.75, figsize=(10, 10));
```
