```python
#função para verificar se possui valores duplicados (isso foi reaproveitado das classes e pode ser bem útil)
def tem_duplicados(df):
    if df.duplicated().any():
        print("Possui {0} valores duplicados".format(df.duplicated().sum()))
    else:
        print("Não possui valores duplicados")        
tem_duplicados(df_titan) #eu entro o nome do meu datafile aqui na chamada da função!
```

3. no nosso caso, observamos um elevado número de *crianças* mortas e um número curioso de *idosos* que foram salvos. Isso normalmente indica *distorções* no que aconteceu de fato naquele navio, no momento do acidente. Algumas vezes, estas distorções são inclusive *criminosas*;
4. além disso, pudemos observar, ao cruzar o número de parentes embarcados com idade do passageiro, uma curiosa tendência a, quando o número de parentes embarcados ser mais elevado, a idade do passageiro tender a ser reduzida. É claro que isso é intuitivo, pois *famílias grandes* tendem a ter *crianças*! Como não estávamos esperando inicialmente encontrar tal correlação, houve a surpresa. É legal explorar dados, algumas vezes somos surpreendidos por alguma *correlação* curiosa (e algumas dessas correlações deram origem por exemplo, à descoberta da [penicilina](https://pt.wikipedia.org/wiki/Penicilina)!
<br>

E, por falar em penicilina, gostaria de citar o trecho da Wikipédia na qual cita que: *"tem-se dito que muitas descobertas científicas são feitas ao acaso. O acaso, já dizia Pasteur, só favorece aos espíritos preparados e não prescinde da observação."*. Lindas palavras. E do meu ponto de vista, nada mais verdadeiro!


**Limitações**

1. em Estatística, normalmente se trabalha com *amostras*. A partir do tratamento dessas amostras e baseados na [lei dos grandes números](https://pt.wikipedia.org/wiki/Lei_dos_grandes_n%C3%BAmeros), nós podemos tirar conclusões baseados na *inferência estatística*;
2. ocorre que em nosso caso, não estamos trabalhando com *amostras* e sim com um conjunto *selecionado* e *não aleatório* de dados. Isso altera totalmente o resultado da análise e ela passa a não ser mais confiável para fins de *inferência*;
3. Embora tenhamos diversos gráficos apontando possíveis e curiosoas distorções nas contagens e dispersão de dados, devemos elmbrar que estes são resultados meramente *ilustrativos*, uma vez que violamos diversos princípios básicos da Estatística;
4. outra coisa que devemos perceber é que estaos trabalhando com, além de uma pequena amostra *não aleatória*, escolhida a fim de pegar os registros menos deteriorados do navio, com um número *limitado* de campos;
5. quando temos poucos campos em um determinado banco de dados, qualquer análise se torna *crítica*, uma vez que não temos como cruzar informações e verificar a *coerência interna* desses dados;
6. para piorar, o número de colunas foi realmente pequeno! Então foi necessário agir com muita criatividade para obter algum *suco* destes dados (ou seja, consultas que nos fornecessem *informação útil*, isso é mais ou menos como espremer uma laranja!);
7. e por fim, uma das melhores maneiras de se trabalhar com *consultas de bancos de dados* é cruzar informações de diversas fontes. Por exemplo, cruzar os dados do banco do Titanic com os dados do [Museu do Holocausto](http://www.yadvashem.org/). O Yad Vashem mantém um banco de dados genealógico de todas as famílias de hebreus. O cruzamento dos nomes dos passageiros poderia em alguns casos, nos trazer informações interessantes sobre possíveis laços de parentescos ocultos, motivo da viagem, entre outros. Este é apenas um exemplo. Mas de fato, não fizemos nenhum cruzamento de dados com fontes *externas*, o que é um grande erro!
